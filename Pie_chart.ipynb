#### 1: List the causes of death and percentile in Ohio state in the year 2012

```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# View the plot 
%matplot inline
```

```python
#cause of deaths
cause=['Chronic Diseases','Unintentional Injuries','Alzheimiers','Influenza and Pneumonia', 'Sepsis','Others']
#percentile
percentile=[62,5,4,2,1,26]
```

#### 2: Draw a pie chart using the two data points

```python
#set the pie chart plot propertie
```

```python
#set figure size
plt.figure(figsize=(20,20))
```

```python
#explode the largest pie in the data set
explode=(0.05,0,0,0,0,0)
```

```python
#set pie chart properties
plt.pie(percentile, labels=cause, explode=explode, autopct='%1.1f%%',startangle=70)
#set axis equal to draw pie as circle
plt.axis('equal')
plt.title('Ohio State 2012 - Leading Causes of Death')
plt.show()
```
